### START TIME

```python
df.Start_Time
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time[0]
```

```python
sns.distplot(df.Start_Time.dt.hour, bins=24, kde=False, norm_hist=True)
```

- A high percentage of accident occur between 4 pm to 6 pm (Probably people leaving offices to go home and may be traffic could be reason)
- Next highest percentage of accidents is between 6 am to 9am (Probably people commuting to thier work and were in a hurry and also the traffic at that time) 


```python
sns.distplot(df.Start_Time.dt.dayofweek, bins=7, kde=False, norm_hist=True)
```

Is the distribution of accidents by hour the same on weekends as on weekdays.

```python
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(sundays_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

```python
Mondays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.distplot(Mondays_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

On Sundays the Peak occurs between 4 pm to 10 pm and then decreases unlike weekdays

```python
sns.distplot(df.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
```

Accidents more likely to occur in Winters (Probably because the lack of street lighting,Fog and smog could be the Major Factors For the road accidents)
